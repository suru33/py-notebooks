```python
import json
import pandas as pd
```

```python
with open('datasets/tll-ams.json', 'r') as f:
    data = json.loads(f.read())
```

```python
itineraries = {i['OutboundLegId']: i for i in data['Itineraries']}
```

```python
legs = {l['Id']: l for l in data['Legs']}
```

```python
final_data = []
for k, v in itineraries.items():
    leg = legs[k]
    d = {
        'mins':leg['Duration'],
        'stops':len(leg['Stops'])
    }
    for p in v['PricingOptions']:
        d['price'] = p['Price']
        final_data.append(d)
```

```python
final_data
```

```python
df = pd.DataFrame(final_data)
```
