### [Date format:  `now +/- y m d`](https://regex101.com/r/BOaLXn/1)

> `?P<group>` **: Python way of grouping**

**This regx limits `years` from `1` to `3`, `months` from `1` to `11` and `dates` from `1` to `59`**

```python
r1 = re.compile(r'^date\(now ?(?P<op>\+|-) ?((?P<y>[1-3])y)? ?((?P<m>[1-9]|1[01])m)? ?((?P<d>[1-9]|[1-5]\d)d)?\)$|^date\(now\)$')
```

```python
r1t1 = r1.search('date(now + 3y 5m 10d)')
r1t1
```

```python
r1t1.groupdict()
```

```python
op = r1t1.group('op')
years = r1t1.group('y')
months = r1t1.group('m')
days = r1t1.group('d')
```

```python
print(f'op: {op}, years {years}, months {months}, days {days}')
```

```python
r1t2 = r1.search('now + 100y')
print(r1t2)
```

### [Check `YYYY-MM-DD` formatted date from year `2000` to `2099`](https://regex101.com/r/7zF7l8/1/)

```python
r2 = re.compile(r'^date\((?P<date>20\d{2}-(?:0\d|1[0-2])-(?:3[01]|[12][0-9]|0[1-9]))\)$')
```

```python
r2t1 = r2.search('date(2019-09-02)')
r2t1
```

```python
r2t1.group('date')
```

```python
Image('/tmp/out.png')
```

```python
r2t2 = r2.search('date(1999-09-02)') # invalid date 1999
print(r2t2)
```

### [env(KEY) pattern](https://regex101.com/r/zUcJcB/1)

```python
r3 = re.compile(r'^env\((?P<key>[a-zA-Z]\w*)\)$')
```

```python
r3t1 = r3.search('env(TEST_KEY)')
r3t1
```

```python
r3t1.group('key')
```

```python
print(r3.search('env(_TEST_KEY)'))
```

```python
print(r3.search('env(TE$ST_KEY)'))
```
