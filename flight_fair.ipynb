```python
import pandas as pd
from IPython.display import display_html
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def dss(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
df = pd.read_pickle('datasets/sky-tll-ams.p')
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df['x'] = df['mins'] * df['price']
```

```python
plt.scatter(df['price'], df['mins'])
plt.xlabel('Price')
plt.ylabel('Mins')
plt.show()
```

```python
dss(df.head(), df.tail())
```

```python
df_zscore = (df - df.mean())/df.std()
```

```python
dss(df.sort_values(by=['mins']), df.sort_values(by=['price']), df.sort_values(by=['x']))
```

```python
dss(df_zscore.sort_values(by=['mins']), df_zscore.sort_values(by=['price']), df_zscore.sort_values(by=['x']))
```

```python
mins = df_zscore['mins'].sort_values()
price = df_zscore['price'].sort_values()
xx = df_zscore['x'].sort_values()
x = range(0, len(mins))
plt.plot(x, mins, c='r', label='mins');
plt.plot(x, price, c='b', label='price');
plt.plot(x, xx, c='g', label='X');
plt.legend(loc='upper left')
plt.show()
```
