```python
import pandas as pd
from IPython.display import display_html
```

```python
def dss(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
df = pd.read_pickle('datasets/sky-final.p')
```

```python
price_df = df.sort_values(by=['price', 'mins', 'stops'])
```

```python
dss(df, price_df)
```

```python
t_idx = list(df.index.values)
p_idx = list(price_df.index.values)
```

```python
t_idx_diff = [abs(t_idx.index(i) - p_idx.index(i)) for i in t_idx]
```

```python
df['diff'] = t_idx_diff
```

```python
df
```

```python
final_df = df.sort_values(by=['diff','price', 'mins', 'stops'])
```

```python
dss(df, final_df)
```
