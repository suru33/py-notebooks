```python
import pandas as pd
from IPython.display import display_html
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
df = pd.read_pickle('datasets/sky-tll-ams.p')
```

```python
df['x'] = df['mins'] * df['price']
```

```python
plt.scatter(df['price'], df['mins'])
plt.xlabel('Price')
plt.ylabel('Mins')
plt.show()
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
display_side_by_side(df.head(), df.tail())
```

```python
df_zscore = (df - df.mean())/df.std()
```

```python
display_side_by_side(df, df_zscore)
```

```python
display_side_by_side(df.sort_values(by=['mins']), df.sort_values(by=['price']), df.sort_values(by=['x']))
```
